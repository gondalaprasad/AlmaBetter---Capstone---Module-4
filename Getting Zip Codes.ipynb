```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load your Excel file into a DataFrame
df = pd.read_excel('/content/Neighbourhoods for Spatial.xlsx')

# Function to fetch zip code for a given place
def get_zip_code(place_column_name):  # Change 'Place' to the actual column name
    url = f'https://www.mapdevelopers.com/what-is-my-zip-code.php?address={place_column_name.replace(" ", "%20")}'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        zip_info = soup.find('div', class_='zip-info')

        if zip_info:
            zip_code = zip_info.get_text(strip=True)
            return zip_code
        else:
            return None
    else:
        return None

# Apply the function to fetch zip codes for each place in the DataFrame
df['Zip Code'] = df['Place'].apply(get_zip_code)  # Change 'Place' to the actual column name

# Display the updated DataFrame
print(df)
```
