```python
#importing necessary libraries
import pandas as pd
```

```python
play_store_data_id = '15FNnUlLDj8fD566CnLYUNdmklkM2qBPl2dTKsO6oS9g'
user_reviews_id = '1-g9J_udlUHjCf02V5Sm0N-h0qFVoVxAKAwG5A61CNM8'
sheet_name = 'Sheet1'

play_store_data = "https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(play_store_data_id, sheet_name)
user_reviews_data ="https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(user_reviews_id, sheet_name)
```

```python
play_store_data_df = pd.read_csv(play_store_data)
play_store_data_df.info()
```

```python
play_store_data_df['Rating_Mean'] = play_store_data_df['Rating'].fillna(play_store_data_df['Rating'].mean())
```

```python
play_store_data_df.describe()
```

```python
play_store_data_df['Rating_CWM'] = play_store_data_df.groupby('Category')['Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
play_store_data_df.describe()
```

```python
# Droping the original 'Rating' column
play_store_data_df.drop('Rating', axis=1, inplace=True)

# Droping the temporary 'Rating_Mean' column
play_store_data_df.drop('Rating_Mean', axis=1, inplace=True)
```

```python
play_store_data_df['Price'] = play_store_data_df['Price'].replace('$','')
play_store_data_df.to_excel('play.xlsx')
```

```python
# Removing duplicates based on the 'App' column
play_store_data_df.drop_duplicates(subset='App', inplace=True)
```

```python
# Drop rows where 'Type' is '0'
play_store_data_df = play_store_data_df[play_store_data_df['Type'] != '0']
play_store_data_df = play_store_data_df.dropna(subset=['Type'])

play_store_data_df = play_store_data_df.dropna(subset=['Current Ver'])

play_store_data_df = play_store_data_df.dropna(subset=['Android Ver'])
```

```python
play_store_data_df.to_csv('play.csv')
```
